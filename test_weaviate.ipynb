# Generate vectors


```python
import requests
import pandas as pd
import os
from typing import List
import openai
from openai import OpenAI

OPENAI_API_KEY = "QjzR8b1Xq2c3Y4xN5z6Z7a8BcD9eF0G"
OPENAI_PROXY_SG = 'http://api-gw.motiong.net:5000/api/openai/eve/v1'

openai.api_key = OPENAI_API_KEY
openai.base_url = OPENAI_PROXY_SG

client = OpenAI(api_key=OPENAI_API_KEY, base_url=OPENAI_PROXY_SG)

def vectorize(texts: List[str], model="text-embedding-3-small") -> List[float]:
    response = client.embeddings.create(
        model=model,
        input=texts
    )

    embeddings = [embedding.embedding for embedding in response.data]
    return embeddings

```

```python
data_url = "https://raw.githubusercontent.com/weaviate-tutorials/edu-datasets/main/movies_data_1990_2024.json"
resp = requests.get(data_url)
df = pd.DataFrame(resp.json())

emb_dfs = list()
src_texts = list()

for i , row in enumerate(df.itertuples(index= False)):
    src_text = f"Title{row.title}; Overview: {row.overview}"
    src_texts.append(src_text)

    if len(src_texts) == 150 or i == len(df) - 1:
        output = vectorize(src_texts)
        index = list(range(i - len(src_texts) + 1, i + 1))
        emb_df = pd.DataFrame(output, index=index)
        emb_dfs.append(emb_df)
        src_texts = list()

emb_df = pd.concat(emb_dfs)

os.makedirs("scratch", exist_ok=True)

emb_df.to_csv("scratch/movies_embeddings.csv", index=False)
```

# Import Data

```python
import weaviate
import pandas as pd
import requests
from datetime import datetime, timezone
import json
from weaviate.util import generate_uuid5
from tqdm import tqdm
import os

data_url = "https://raw.githubusercontent.com/weaviate-tutorials/edu-datasets/main/movies_data_1990_2024.json"
data_resp = requests.get(data_url)
df = pd.DataFrame(data_resp.json())

embs_path = "https://raw.githubusercontent.com/weaviate-tutorials/edu-datasets/main/movies_data_1990_2024_embeddings.csv"

embs_df = pd.read_csv(embs_path)

movies = client.collections.get("MovieCustomVector")

with movies.batch.dynamic() as batch:
    for i, movie in enumerate(df.itertuples(index=False)):
        release_date = datetime.strptime(movie.release_date, "%Y-%m-%d").replace(
            tzinfo=timezone.utc)
        
        genre_ids = json.loads(movie.genre_ids)

        movie_obj = {
            "title": movie.title,
            "overview": movie.overview,
            "vote_average": movie.vote_average,
            "genre_ids": genre_ids,
            "release_date": release_date,
            "tmdb_id": movie.id
        }

        vector = emb_df.iloc[i].values.tolist()

        batch.add_object(
            properties = movie_obj, 
            uuid = generate_uuid5(movie.id),
            vector = vector
        )

if len(movies.batch.failed_objects) > 0:
    print(f"Failed to import {len(movies.batch.failed_objects)} objects")

client.close()    
```

# Vector Search

```python
import weaviate
import weaviate.classes.query as wq
import os

from typing import List
import openai
from openai import OpenAI

OPENAI_API_KEY = "QjzR8b1Xq2c3Y4xN5z6Z7a8BcD9eF0G"
OPENAI_PROXY_SG = 'http://api-gw.motiong.net:5000/api/openai/eve/v1'

openai_client = OpenAI(api_key=OPENAI_API_KEY, base_url=OPENAI_PROXY_SG)

client = weaviate.connect_to_local()

def vectorize(texts: List[str], model="text-embedding-3-small") -> List[List[float]]:
    response = openai_client.embeddings.create(
        model=model,
        input=texts
    )

    embeddings = [embedding.embedding for embedding in response.data]
    return embeddings

query_text = "dystopian future"
query_vector = vectorize([query_text])[0]

movies = client.collections.get("MovieCustomVector")

response = movies.query.near_vector(
    near_vector = query_vector,
    limit = 5,
    return_metadata=wq.MetadataQuery(distance=True)
)

for o in response.objects:
    print(
        f"{o.properties['title']} {o.properties['release_date'].year}")
    
    print(f"Distance to query: {o.metadata.distance:.3f}\n")

client.close()    
```

```python
client = weaviate.connect_to_local()
movies = client.collections.get("MovieCustomVector")

response = movies.query.bm25(
    query="history", limit=5, return_metadata=wq.MetadataQuery(score=True)
)

for o in response.objects:
    print(f"{o.properties['title']} {o.properties['release_date'].year}")
    print(f"BM25 score: {o.metadata.score:.3f}\n")

client.close()
```

```python
client = weaviate.connect_to_local()
movies = client.collections.get("MovieCustomVector")

query = "history"
query_vector = vectorize([query])[0]

response = movies.query.hybrid(
    query="history", 
    vector=query_vector, 
    limit=5, 
    return_metadata=wq.MetadataQuery(score=True),
)

for o in response.objects:
    print(f"{o.properties['title']} {o.properties['release_date'].year}")
    print(f"Hybrid score: {o.metadata.score:.3f}\n")

client.close()
```

```python
client = weaviate.connect_to_local()

movies = client.collections.get("MovieCustomVector")

response = movies.query.near_vector(
    near_vector = query_vector,
    limit = 5,
    return_metadata=wq.MetadataQuery(distance=True),
    filters=wq.Filter.by_property("release_date").greater_than(datetime(2020, 1, 1))
)

for o in response.objects:
    print(
        f"{o.properties['title']} {o.properties['release_date'].year}")
    
    print(f"Distance to query: {o.metadata.distance:.3f}\n")

client.close()
```
